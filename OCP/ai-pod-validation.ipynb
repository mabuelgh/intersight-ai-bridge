```python
from langchain_nvidia_ai_endpoints import ChatNVIDIA

llm = ChatNVIDIA(base_url="http://meta-llama3-1b-instruct-openshift-operators.apps.CLUSTER_NAME.BASE_DOMAIN/v1", model="meta/llama-3.2-1b-instruct",nvidia_api_key="not-used", max_tokens=1024)
```

```python
LLM_PROMPT_TEMPLATE = (
    "<s>[INST] <<SYS>>"
    "{system_prompt}"
    "{input}"
    "<</SYS>>"
)
GENERATE_PROMPT="You are an AI assistant. Please answer the user queries appropriately."
```

```python
# Generate LLM prompt
prompt = LLM_PROMPT_TEMPLATE.format(system_prompt=GENERATE_PROMPT,input="tell me about France")
# Generate instruction for the context
response = llm.invoke(prompt).content
print(response)
```
